```python
import pandas as pd
```

```python
class MissingValueHandler:
  def __init__(self, dataframe):
    self.dataframe = dataframe

  def fill_with_mean(self, columns):
    for col in columns:
      if col in self.dataframe.columns:
        self.dataframe[col].fillna(self.dataframe[col].mean(), inplace = True)

  def fill_with_mode(self, columns):
    for col in columns:
      if col in self.dataframe.columns:
        self.dataframe[col].fillna(self.dataframe[col].mode().iloc[0], inplace = True)

  def fill_with_median(self, columns):
    for col in columns:
      if col in self.dataframe.columns:
        self.dataframe[col].fillna(self.dataframe[col].median(), inplace = True)




  def remove_rows(self):
    self.dataframe.dropna(axis = 0, inplace = True)

  def get_dataframe(self):
    return self.dataframe
```

```python
data = {'A': [1,2,3, None, 5],
        'B': [None, 2, 3, 4, 5],
        'C':[1,2,3,4,5],
        'D':['X', 'Y', 'Z', None, 'Z']}
df = pd.DataFrame(data)
print(df)
missinghandler1 = MissingValueHandler(df)
missinghandler1.fill_with_mode(['D'])
updated_df =missinghandler1.get_dataframe()
updated_df
```
